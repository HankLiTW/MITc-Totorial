## 一場考試，如果分數低於40分，就死當，40~60，可以補考到60分，60以上就通過。用if elif else寫一個函數去做判斷。

```python
def grade(score):
    if score<40:
        print("死當")
    elif score<60:
        print("補考")
    else:
        print("通過")
```

```python
[grade(20),grade(40),grade(60)]
```

## 找出1~1000的質數

### 你需要去思考：1.質數的性質 2.可能程式架構
#### 質數的性質為"該數因數只有自己與1"
#### 可能程式架構為 for 和 while ，因為你必須對多個數字做重複判斷。
##### 當然你也可以用function和if 寫一個判斷函數，然後一個一個塞進去，最後print出來一個list包含所有的質數。

```python
prime = []
for number in range(2,1000):
    count = 0
    for j in range(1,1000):
        if number%j == 0 :
            count += 1
    if count==2 :
        prime.append(number)
print(prime)
```

## 找出1~1000中所有7的倍數（用While解題）

```python
number = 1
seven = []
while number<=1000:
    if number%7 == 0:
        seven.append(number)
    number += 1
print(seven)
```

### While True 版（不推薦）

```python
number = 1
seven = []
while True:
    if number%7 == 0:
        seven.append(number)
    number += 1
    if number>1000:
        break
print(seven)
```

## 找出1~1000的質數（用While解題）

```python
prime = []
number = 1
while number <= 1000:
    count = 0
    number += 1
    j = 1
    while j <= 1000:
        if number%j == 0 :
            count += 1
        j += 1
    if count==2 :
        prime.append(number)
print(prime)
```

## 試求 $ -x^2+2x+20 + log(x^2) = 0 $ 的最大值，並且找到能使此方程式最大值最小的x

```python
import numpy as np
```

```python
def f(x):
    return(-x**2+2*x +20 + np.log(x**2) )
f_temp = 0
min_x = 0
for i in range(1,10):
    if f(i) > f_temp:
        f_temp = f(i)
        min_x = i
print(min_x,f_temp)
```

## 給定 $ P(X) = \frac{\lambda^x e^-\lambda}{x!} , where \lambda = 3 , 0<=x<=10 則 E(X)=?$

```python
def factori(n) :
    fac_list = list(range(1,n+1))
    return(np.prod(fac_list))
def prob(x):
    lambd = 3
    return(lambd**x*np.exp(-lambd)/(factori(x)))
exp= 0
for i in range(1,10):
    exp += i*prob(i)
print(exp)
```

### 基本統計量

```python
# 平均（根據大數法則，當取很多點的時候，平均會趨近於期望值）
x = np.random.randn(1000)
print(np.mean(x))
print(x.mean())
```

```python
#變異數
print(np.var(x))
print(x.var())
```

```python
#標準差
print(np.std(x))
print(x.std())
```

```python
#標準化
x = np.random.randn(1000)*10
x_new = (x-x.mean())/x.std()
print(x_new.mean())
print(x_new.std())
```

```python
#最大最小值和中位數
x = np.random.randn(1000)*10
print(x.max())
print(x.min())
print(np.median(x))
```

```python
#當然也可以反向標準化
x = np.random.randn(1000)
#我們想要創造一個N(100,10)的樣本
x_new = x*10+100
print(x_new.mean())
print(x_new.std())
```
