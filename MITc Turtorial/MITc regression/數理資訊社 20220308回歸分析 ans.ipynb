```python
import statsmodels.api as sm
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

# Q1:Try to establish a regression models, $\beta_0 = 20$ $\beta_1 = 10$ and x is from normal(0,1) and check assumptions

```python
beta_0 = 20
beta_1 = 10
x = np.random.randn(1000)
e = np.random.randn(1000)
```

```python
y = beta_0+beta_1*x+e
```

```python
x_constant = sm.add_constant(x)
results = sm.OLS(y,x_constant).fit()
results.summary()
```

```python
#assumption1
e_fitted = results.resid
print(np.mean(e_fitted))
print(np.var(e_fitted))
plt.hist(e_fitted)
```

```python
#assumption2
print(np.cov(x,e_fitted))
```

```python
#assumption3
plt.hist(x)
```

```python
#assumption4
y_fitted = results.fittedvalues
plt.hist(y_fitted)
```

# $Q_2$: establish a multiple regression model with $ \beta_0 = 10 \beta_1 = 20 \beta_2 = 30 $,  $x_1$ and $ x_2$ are from N(0,1)

```python
beta_0 = 20
beta_1 = 10
beta_2 = 30
x_1 = np.random.randn(1000)
x_2 = np.random.randn(1000)
e = np.random.randn(1000)
x = np.column_stack((x_1,x_2))
```

```python
y = beta_0 + beta_1*x_1 + beta_2*x_2 + e
```

```python
x_constant = sm.add_constant(x)
results = sm.OLS(y,x_constant).fit()
results.summary()
```
