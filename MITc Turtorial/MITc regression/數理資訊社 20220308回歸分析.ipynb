# 1. 簡單線性迴歸模型

# Our model is $y = \beta_0 + \beta_1 * x + \epsilon $

```python
#create x
x = np.random.randn(1000)*10
```

```python
plt.plot(x)
```

```python
plt.hist(x)
```

# create y  $\beta_0$ = 50 $\beta_1$ = 100


```python
y = 100*x+50
```

# Now, Our model y = $\beta_0$+$\beta_1$*$x$, and we lost $\epsilon$


# Remember that $\epsilon$ is from Normal(0,1) in our assumption

```python
#create error

error = np.random.randn(len(y))
```

```python
y = y+error
```

# The model has been established, we are going to verify our model.

```python
print(x)
```

```python
x_constant = sm.add_constant(x)
```

```python
print(x_constant)
```

```python
#Use LSE to estimate $\beta_0$ and $\beta_1%
results = sm.OLS(y,x_constant).fit()
```

```python
#Check you model
print(results.summary())
```

```python
results.rsquared
```

```python
results.params
```

# It is close to $\beta_0$ and $\beta_1$

```python
y_fitted = results.fittedvalues
```

```python
y_fitted
```

```python
plt.plot(x, y, 'o')
plt.plot(x,y_fitted)
```

```python
e = y-y_fitted 
```

```python
plt.plot(e)
```

```python
plt.hist(e)
```

```python
np.mean(e)
```

```python
np.var(e)
```

# it is seem to be normal distribution, which statisfies our assumption on error

# 2. 多元線性迴歸模型

# Our model is $ y = \beta_0 + \beta_1 * x_1 + \beta_2 *x_2 +\epsilon $

```python
#create x1 x2
x1 =np.random.randn(1000)*10
x2 =np.random.randn(1000)*20
```

```python
#create y 
y = x1+x2+20+np.random.randn(1000)
```

# Only use $x_1$ 

# now our model is  $ y = \beta_0 + \beta_1 * x_1+\epsilon $

```python
x_constant = sm.add_constant(x1)
```

```python
results = sm.OLS(y,x_constant).fit()
print(results.summary())
```

```python
y_fitted = results.fittedvalues
```

```python
plt.plot(x1,y_fitted)
plt.plot(x1,y,'o')
```

# We cannot find accurate relations between $x_1$ and $y$
# Thus, we guess $x_1$ cannot provide enough imformation for $y$
# Now, our model is $ y = \beta_0 + \beta_1 * x_1 + \beta_2 *x_2 +\epsilon $

```python
X = np.column_stack((x1,x2))
```

```python
x1
```

```python
x2
```

```python
X
```

```python
X_constant = sm.add_constant(X)
```

```python
results = sm.OLS(y,X_constant).fit()
print(results.summary())
```

```python
y_fitted = results.fittedvalues
```

```python
e = y - y_fitted
```

```python
plt.plot(e)
```

# Multiple linear case

```python
x1 = np.random.randn(1000)*20
x2 = np.random.randint(0,10,1000)*1/2
x3 = np.random.randn(1000)*23
x4 = np.random.normal(0,2,1000)
```

```python
y = 10+2*x1+3*x2+4*x4+np.random.randn(1000)
```

# Suppose we have to models
# $model_1$ is $y = b_0 +b_1*x_1+b_2*x_2+b_3*x_3$
# $model_2$ is $y = b_0 +b_1*x_1+b_2*x_2+b_3*x_4$
# which one is better?

# Step1, plot $X_i$ and $y$ 

```python
plt.scatter(x1,y)
```

```python
plt.scatter(x2,y)
```

```python
plt.scatter(x3,y)
```

```python
plt.scatter(x4,y)
```

# We cannot find some useful imformation about selection

# Step2, do the regression 

# $model_1$  $y = b_0 +b_1*x_1+b_2*x_2+b_3*x_3$

```python
X = np.column_stack((x1,x2,x3))
```

```python
X_constant = sm.add_constant(X)
```

```python
results = sm.OLS(y,X_constant).fit()
```

```python
print(results.summary())
```

```python
results.rsquared_adj
```

```python
results.mse_total
```

# $model_2$  $y = b_0 +b_1*x_1+b_2*x_2+b_3*x_4$

```python
X = np.column_stack((x1,x2,x4))
```

```python
X_constant = sm.add_constant(X)
```

```python
results = sm.OLS(y,X_constant).fit()
```

```python
print(results.summary())
```

```python
results.rsquared_adj
```

```python
results.mse_total
```

# Step3 :Seems $model_2 $ is better , check its residual plot
