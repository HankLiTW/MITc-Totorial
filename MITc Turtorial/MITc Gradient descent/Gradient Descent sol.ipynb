### Q : 試試看用Adam找找看g(x)的最小值，$\alpha$ = 0,9 ,$\beta$ = 0.999 , learning_rate = 0.001

```python
from sympy import *
## f(x)=x**2-2*x+6
x = symbols("x")
def f(x):
    return(x**2-2*x+6)

f_prime = diff(f(x),x)
x_new = 0
learning_rate = 0.001
## 終止條件為微分小於10**(-5)
x_new = 0
V=0
N=0
alpha = 0.9
beta = 0.999
count=0
while abs(f_prime.evalf(subs = {"x":x_new})) >= 10**(-5):
    count += 1 
    V = alpha*V+(1-alpha)*f_prime.evalf(subs = {"x":x_new})
    N = beta*N+(1-beta)*(f_prime.evalf(subs = {"x":x_new}))**2
    V_hat = V/(1-alpha**(count))
    N_hat = N/(1-beta**(count))
    x_new = x_new - learning_rate*V_hat/sqrt(N_hat+0.0001)
print(f(x_new))
print(count)
```

### Q: 試著解看看以下聯立
Solve

$$3x + 4y + 6z = 10$$
$$ x + y + 3z= 5 $$
$$ 5x+3y+2z=15$$

```python
from scipy.optimize import fsolve
```

```python
def h(X):
    x=X[0]
    y=X[1]
    z=X[2]
    eq0 = 3*x+4*y+6*z-1
    eq1 = x+y+3*z-5
    eq2 = 5*x+3*y+2*z-15
    return eq0,eq1,eq2
```

```python
print(fsolve(h,[0,0,0]))
ans = fsolve(h,[0,0,0])
```
